```python
### para modificar los archivos del INKSCAPE para que los modifique QGIS
import os
import glob
```

```python
#Bienvenida
print("Bienvenido a SVG INKSCAPE-QGIS converter 1.0")
```

```python
cwd = os.getcwd()

for file in glob.glob(cwd + "\\*.svg"):
    
    with open(file, 'r') as f:        
        #archivo de salida con formato para QGIS
        wf = open (f.name.replace(".svg","-QGIS.svg"),'w')
        for line in f:
            wf.write(line)
            if line[:9] == '    <path':
                wf.write('       fill="param(fill) #fff"' + '\n') 
                wf.write('       stroke="param(outline) #000"' + '\n') 
                wf.write('       stroke-width="param(outline-width) 0.3"' + '\n')
                wf.write('       opacity="param(opacity)"' + '\n')
    wf.close()
```
