```python
import matplotlib.pyplot as plt
```

```python
e1o = 3.4
e2o = 3.4
tita0 = 90
```

```python
tita1 = [85, 80, 74, 69, 65, 61, 56, 48, 45, 41, 33, 29, 25, 22, 16]
omega, psi = [],[]

for i in range(len(tita1)):
    omega.append(abs(tita1[i]-tita0))  
    psi.append(abs(tita1[i]-tita0))  

```

```python
psi
```

```python
#Cálculo elongación
d1 = [3.4,3.4,3.4,3.4,3.4,3.4,3.4,3.4,3.4,3.4,3.4,3.4,3.4,3.4,3.4]
d2 = [3.5,3.55,3.6,3.7,3.8,4.0,4.2,4.5,4.8,5.2,6.0,7.0,8.4,10.2,14.0]
e1 , e2 = [], []

for i in range(len(d1)):
    e1.append((d1[i] - e1o)/e1o)  
    e2.append((d2[i] - e2o)/e2o)
```

```python
#Cálculo estiramiento
s1,s2,r1,delta=[],[],[],[]
for i in range(len(d1)):
    s1.append(1+e1[i])
    s2.append(1+e2[i])
    r1.append(s1[i]/s2[i])
    delta.append((s1[i]*s2[i])-1)
```

```python
#Cálculo Razón axial
r1 = s1/s2
```

```python
titaP = tita1 
```

```python
delta = (s1*s2)-1
```

```python
len(s2)
```

```python
for i in range(len(d1)):
    print("W: {}, Psi: {}, e1: {}, e2: {}, s1: 1.0, s2: {}, rs: {}, titaP: {}, delta: {}"
          .format(omega[i], psi[i], e1[i], round(e2[i], 2), round(s2[i],2), round(r1[i],2), tita1[i], round(delta[i],2)))
```

```python
plt.rcParams["figure.figsize"] = (12,12)

plt.scatter(x=psi, y=r1, marker='^',zorder=1)
plt.scatter(x=psi, y=s1, marker='o',zorder=2)
plt.scatter(x=psi, y=s2, marker='d',zorder=3)


#plot = gdfarea5.plot(kind='bar', width = 0.6,zorder=2, legend=None)
# Turn on the grid

plt.minorticks_on()
plt.grid(linestyle='solid',zorder=0)
#plt.grid(which='major', linestyle=':', linewidth='0.2', color='green')
#plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

plt.xticks(fontsize=10,rotation=0)
plt.yticks(fontsize=10,rotation=0)
plt.title("Razón Axial, S1 y S2",fontsize=25,fontweight=300)
plt.xlabel("Ángulo de cizalla",fontsize=15,rotation=0)

plt.ylabel("Parámetros",fontsize=15,rotation=90)
#plt.grid()
plt.legend(["Razón Axial","S1","S2"],loc = "center",bbox_to_anchor=(0.25, 0.85), shadow=True,fontsize=20)

#plt.bar(dic.keys(), dic.values([0][0]))
plt.savefig('R1.png')
plt.show()
```
